```python
name_file_ar = 'content/lastfm_artist_list.csv'
name_file_us = 'content/lastfm_user_scrobbles.csv'
```

```python
import pandas as pd
```

```python
data_1 = pd.read_csv(name_file_ar)
```

```python
data_1
```

```python
data_2 = pd.read_csv(name_file_us)
```

```python
data_2
```

```python
data = pd.merge(data_2, data_1, on='artist_id')
```

```python
data
```

```python
data = data.drop(['artist_id'], axis='columns')
```

```python
data
```

```python
import numpy as np
import sklearn
from sklearn.decomposition import TruncatedSVD
```

```python
recommend_data = data.pivot_table(values='scrobbles', index='user_id', columns='artist_name', fill_value=0)
```

```python
X = recommend_data.T
```

```python
SVD = TruncatedSVD(n_components=12, random_state=5)
resultant_matrix = SVD.fit_transform(X)
resultant_matrix.shape
```

```python
corr_mat = np.corrcoef(resultant_matrix)
corr_mat.shape
```

```python
col_idx = recommend_data.columns.get_loc("Nyktalgia")
corr_specific = corr_mat[col_idx]
pd.DataFrame({'corr_specific':corr_specific, 'Artist': recommend_data.columns})\
.sort_values('corr_specific', ascending=False)\
.head(10)
```
