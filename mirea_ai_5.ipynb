```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas
from IPython.display import display
```

```python
url = 'https://memepedia.ru/memoteka/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml') # достаем html код
print(soup)
```

```python
main_inf = [] # список с главной информацией по мемам
mem_inf = [] # список с информацией по промежуточному мему, добавляется в main_inf
category = [] # список категорий, который свяжется со списком main_inf
```

```python
def time_meme(ur):
  """функция выдает время, когда был загружен мем.
  Передается ссылка на мем. Из тэга time достается время и выводится"""
  response = requests.get(ur)
  soup = BeautifulSoup(response.text, 'lxml') # достаем html код
  time_find = soup.find("time", class_ = "entry-date published ").text
  return time_find

def views_meme(ur):
  """Функция выдает количество просмотров мема.
  Передается ссылка на мем. Из тэга span достаются просмотры и выводятся"""
  response = requests.get(ur)
  soup = BeautifulSoup(response.text, 'lxml') # достаем html код
  views_find = soup.find("span", class_ = "count").text
  return views_find

```

```python
links_img_find = soup.find(class_ = "post-items").find_all("div", class_ = "post-thumbnail") # ищем в классе постов на основной странице другие классы и названия мемов
for item in links_img_find:
  mem_inf = [] # обнуляем данные, чтобы недозаписывать
  url_meme = item.find("a").get("href") # достаем ссылку на мем
  name_meme = item.find("a").get("title") # достаем имя мема
  mem_inf.append(name_meme) # добавляем имя мема в информацию о нем
  #time_mem = time_meme(url_meme) # достаем дату загрузки мема
  #mem_inf.append(time_mem) # добавляем время мема
  #views_mem = views_meme(url_meme) # достаем количество просмотров
  #mem_inf.append(views_mem) # добавляем количество просмотров мема
  main_inf.append(mem_inf) # добавляем список информации по мему в главный список информации о мемах

category_find = soup.find(class_ ="post-items").find_all("div", class_ = "content")
for item in category_find:
  category_meme = item.find_all("a", rel = "category tag") # достаем все категорию
  if len(category_meme) > 1: # категорий бывае 1 или 2, поэтому сравниваем размер списка категорий
    category_name = category_meme[0].text + " и " + category_meme[1].text # итоговое имя категорий
  else:
    category_name = category_meme[0].text
  category.append(category_name)
print(category)
print(main_inf)
for i in range(len(main_inf)):
  main_inf[i].append(category[i])
print(main_inf)


```

```python
with open('data_test.csv', 'w', newline='') as f:
  writer = csv.writer(f)
  name_columns = ['название', 'дата', 'просмотры', 'категории']
  writer.writerow(name_columns)
  writer.writerows(main_inf)
itog = pandas.read_csv('data_test.csv', delimiter=';')
print(itog)
f.close()
```

```python
for i in range(2, 124): # проходимся по всем страницам
  mem_inf = [] # список с информацией по промежуточному мему, добавляется в main_inf
  category = [] # список категорий, который свяжется со списком main_inf
  url = 'https://memepedia.ru/memoteka/'
  url = url + 'page/' + str(i) + '/' 
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'lxml') # достаем html код
  links_img_find = soup.find(class_ = "post-items").find_all("div", class_ = "post-thumbnail") # ищем в классе постов на основной странице другие классы и названия мемов
  for item in links_img_find:
    mem_inf = [] # обнуляем данные, чтобы недозаписывать
    url_meme = item.find("a").get("href") # достаем ссылку на мем
    name_meme = item.find("a").get("title") # достаем имя мема
    mem_inf.append(name_meme) # добавляем имя мема в информацию о нем
    time_mem = time_meme(url_meme) # достаем дату загрузки мема, стопорит парсинг(переход на другую страницу)
    mem_inf.append(time_mem) # добавляем время мема
    views_mem = views_meme(url_meme) # достаем количество просмотров, стопорит парсинг(перехоод на другую страницу)
    mem_inf.append(views_mem) # добавляем количество просмотров мема
    main_inf.append(mem_inf) # добавляем список информации по мему в главный список информации о мемах

  category_find = soup.find(class_ ="post-items").find_all("div", class_ = "content")
  for item in category_find:
    category_meme = item.find_all("a", rel = "category tag") # достаем все категорию
    if len(category_meme) > 1: # категорий бывае 1 или 2, поэтому сравниваем размер списка категорий
      category_name = category_meme[0].text + " и " + category_meme[1].text # итоговое имя категорий
    else:
      category_name = category_meme[0].text
    category.append(category_name)
  k = len(main_inf)
  for j in range(k - len(category), k):
    main_inf[j].append(category[j - (k - len(category))])
```

```python
lines = pandas.DataFrame(main_inf)
lines.to_csv('data_test.csv') # запись в файл data_test.csv
itog = pandas.read_csv('data_test.csv')
itog = itog.rename(columns={'Unnamed: 0': 'index', '0': 'name', '1': 'date', '2': 'views', '3': 'category'}) # причесываем данные
itog = itog.drop(columns={'4', '5'}) # удаляем ненужные колонки
itog.to_csv('data_test.csv') # перезапись новой базы в файл
pandas.options.display.max_columns = None # для корректного отображения
display(itog)
f.close()
#rea = pandas.read_csv('data_test.csv')
#print(rea)
```
